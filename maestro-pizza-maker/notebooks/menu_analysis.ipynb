```python
# Compute Taste at Risk (TaR) for each pizza in the menu and for the whole menu and briefly comment on the results
# Use quantile = 0.1
# Answer following question:
# 1. Does the sum of TaR for each pizza in the menu equal to the TaR of the whole menu and why?

# Hint: use functions that you have implemented in the taste_at_risk.py

TaRs = {} 
q = 0.1
for pizza in pizza_menu.pizzas:
    TaRs[pizza.name] = taste_at_risk_pizza(pizza=pizza, quantile=q)
print("TaR of each pizza:")
for name, val in zip(TaRs.keys(),TaRs.values()):
    print(name + ": " + str(val))
print("")
sum_of_TaRs = sum(TaRs.values())
print("Sum of TaR for each Pizza in the Manu: ", sum_of_TaRs)
print("")
menu_TaR = taste_at_risk_menu(menu=pizza_menu, quantile=q)
print("TaR for Menu:", menu_TaR)
print()
print("Is TaR menu > sum of TaRs:", menu_TaR > sum_of_TaRs)
if menu_TaR > sum_of_TaRs:
    print("The TaR of the Menu is larger than the sum of the Pizzas TaR. As the VaR measure, TaR does not satisfy the subadditivity porperty of a coherent risk measure ( no diversification effect )")
```

```python
# Compute Conditional Taste at Risk (CTaR) for each pizza in the menu and for the whole menu and briefly comment on the results

# Hint: use functions that you have implemented in the taste_at_risk.py

C_TaRs = {} 
q = 0.1
for pizza in pizza_menu.pizzas:
    C_TaRs[pizza.name] = conditional_taste_at_risk_pizza(pizza=pizza, quantile=q)
print("C_TaRs of each pizza:")
for name, val in zip(C_TaRs.keys(),C_TaRs.values()):
    print(name + ": " + str(val))
print("")
print("Is CTaR always < than TaR?")
print([
    conditional_taste_at_risk_pizza(pizza=pizza, quantile=q) <= taste_at_risk_pizza(pizza=pizza, quantile=q) for pizza in pizza_menu.pizzas
    ])
sum_of_CTaRs = sum(C_TaRs.values())
print("")
print("Sum of CTaR for each Pizza in the Manu: ", sum_of_CTaRs)
print("")
menu_CTaR = conditional_taste_at_risk_menu(menu=pizza_menu, quantile=q)
print("CTaR for Menu:", menu_CTaR)

# 1) CTaR of Pizza is less than or equal to individual TaR for each Pizza. Same for menu.
# 2) CTaR is a more conservative risk measure than TaR.
```

```python
# Define constraints_values and constraints_ingredients and find the optimal pizza using the func maximize_taste_penalty_price, that you have implemented in the pizza_optimizer.py
# Use lambda_param = 0.5

# Add this pizza to the menu using the add_pizza method of the PizzaMenu class
from maestro_pizza_maker.pizza_optimizer import maximize_taste_penalty_price, PizzaConstraintsValues, PizzaConstraintsIngredients

opt_con_ing = PizzaConstraintsIngredients(
    meat=1,
    cheese=2,
    fruits=0,
    vegetables=1
)

opt_pizza = maximize_taste_penalty_price(PizzaConstraintsValues, opt_con_ing, 0.5)
```

```python
# Compute the new TaR and CTaR for the whole new menu with added pizza

pizza_menu.add_pizza(opt_pizza)
new_TaR_menu = taste_at_risk_menu(pizza_menu, 0.1)
new_C_TaR_menu = conditional_taste_at_risk_menu(pizza_menu, 0.1)

print(f"new TaR manu: {new_TaR_menu}")
print(f"new CTaR menu: {new_C_TaR_menu}")

pizza_menu.remove_pizza(pizza=opt_pizza)
```
